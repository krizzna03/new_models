```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
numeric = df.select_dtypes(include = [np.number])
if numeric.shape[1] >=4:
  plt.figure(figsize= (15,8))
  corr = numeric.corr()
  sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
  plt.title('Correlation')
  plt.tight_layout()
  plt.show()
else:
  print("Not enought numeric columns")
```

```python
df_sorted = df.sort_values('total_cases', ascending=False).head(10)

plt.figure(figsize=(10, 8))
sns.barplot(x='total_cases', y='country', data=df_sorted)
plt.title('Top 10 Countries by Total Cases')
plt.xlabel('Total Cases')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
# Create a pair plot of some key numeric features to observe pairwise relationships
cols_for_pairplot = ['active_cases', 'critical_cases', 'total_cases', 'total_deaths']
available_cols = [col for col in cols_for_pairplot if col in df.columns]
if len(available_cols) >= 2:
    sns.pairplot(df[available_cols].dropna())
    plt.suptitle('Pair Plot of Selected Features', y=1.02)
    plt.show()
else:
    print('Not enough columns for a pair plot.')
```

```python
for col in numeric.columns:
    plt.figure(figsize=(8,4))
    sns.histplot(df[col].dropna(), kde=True)
    plt.title(f'Distribution of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
    plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
target = 'total_deaths'
features = ['active_cases', 'critical_cases', 'total_cases', 'recovered_cases', 'population']
features = [col for col in features if col in df.columns]

if target in df.columns and len(features) > 0:
  model_df = df[features +[target]].dropna()
  x = model_df[features]
  y = model_df[target]
  x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state =42)
  model = LinearRegression()
  model.fit(x_train, y_train)

  y_pred = model.predict(x_test)
  mse = mean_squared_error(y_test, y_pred)
  r2 = r2_score(y_test, y_pred)
  print(f'Mean Squared Error: {mse}')
  print(f'R-squared: {r2}')
else:
  print("ERROR")
```

```python
rmse = np.sqrt(mse)
rmse
```
