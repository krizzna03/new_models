```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.compose import ColumnTransformer
from sklearn.ensemble import RandomForestRegressor
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
num_cols = ['duration', 'days_left']
cat_cols = ['airline', 'source_city', 'departure_time', 'arrival_time', 'destination_city', 'class']
num_transform = Pipeline(steps = [
    ('scaler', StandardScaler())
])
cat_transform = Pipeline(steps =[
    ('encoder', OneHotEncoder(handle_unknown = 'ignore'))
])
preprocess = ColumnTransformer(transformers=[
    ('num', num_transform, num_cols),
    ('cat', cat_transform, cat_cols)
])
forest_model = Pipeline(steps=[
    ('preprocess', preprocess),
    ('forest', RandomForestRegressor(n_estimators = 100, random_state = 42))
])
linear_model = Pipeline(steps=[
    ('preprocess', preprocess),
    ('linear', LinearRegression())
])
```

```python
x = df.drop(columns = ['price', 'flight'], axis = 1)
y = df['price']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.20, random_state = 42)
```

```python
forest_model.fit(x_train, y_train)
linear_model.fit(x_train, y_train)
```

```python
y_rf = forest_model.predict(x_test)
y_lr = linear_model.predict(x_test)
```

```python
print("Random Forest")
print("MSE:", mean_squared_error(y_test, y_rf))
print("MAE:", mean_absolute_error(y_test, y_rf))
print("R2:", r2_score(y_test, y_rf))
```

```python
print("Linear Regression")
print("MSE:", mean_squared_error(y_test, y_lr))
print("MAE:", mean_absolute_error(y_test, y_lr))
print("R2:", r2_score(y_test, y_lr))
```
